```python
import os
import csv

import turtle as t
```

```python
training_path = './data/train_simplified/'
training_files = []
for path in os.listdir(training_path):
    print(path)
    if path.endswith('.csv'):
        training_files.append(path)
```

```python
images = []

with open(training_path + training_files[0]) as csvfile:
    airplane = csv.reader(csvfile)
    for row in airplane:
        images.append(row)
```

```python
r = images[4][1]
```

```python
r = eval(r)
```

```python
for stroke in r:
    t.penup()
    all_x = stroke[0]
    all_y = stroke[1]
    for i in range(len(stroke[0])-1):
        t.goto(all_x[i], all_y[i])
        t.pendown()
    t.goto(all_x[-1], all_y[-1])
    t.penup()
t.done()
```
